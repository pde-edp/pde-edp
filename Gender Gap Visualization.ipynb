# Gender Gap in Several Sectors 

The purpose of this is to visualize the gender gap between and women from 1970 to 2011 in a variety of fields
A gender gap is a difference between men and women with respect to some type of measure or metric. Most of us have heard of the gender gap in pay or in political empowerment. 
This project concerns the gender gap in college degrees conferred in different fields.




```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

women_degrees = pd.read_csv(r"\percent-bachelors-degrees-women-usa.csv")

```

```python
#We split the columns into 3 categories or clusters. 
#The first contains majors that can be broadly classified as belonging to 
#STEM (Science Math, Engineering, and Technology). The second consists of fields more or less considered liberal arts and
#the third we call others


stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 
             'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences',
             'Computer Science', 'Engineering', 'Architecture']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', ]

cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

fig = plt.figure(figsize =(18, 20))

# Generate first column of charts. STEM degrees
for sp in range(0, 6):
    # Going to have 17 plots total, so we'll use a 6 rows, 3 columns format for our plots.
    ax = fig.add_subplot(6, 3, 3 * sp + 1)
    
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c = cb_dark_blue, linewidth = 3, label = 'Women')
    ax.plot(women_degrees['Year'], 100 - women_degrees[stem_cats[sp]], c = cb_orange, linewidth = 3, label = 'Men')
```

```python
# Set plot line colors to colorblind mode
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

fig = plt.figure(figsize =(18, 20))

# Generate first column of charts. STEM degrees
for sp in range(0, 6):
    # Going to have 17 plots total, so we'll use a 6 rows, 3 columns format for our plots.
    ax = fig.add_subplot(6, 3, 3 * sp + 1)
    
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c = cb_dark_blue, linewidth = 3, label = 'Women')
    ax.plot(women_degrees['Year'], 100 - women_degrees[stem_cats[sp]], c = cb_orange, linewidth = 3, label = 'Men')
    
    # Remove the spines from the plots
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    
    # Remove the tick marks from the plot
    ax.tick_params(bottom = False, top = False, left = False, right = False)
    
    # Enable x-axis labels for the bottommost line chart in the STEM columns
    if sp == 5:
        ax.tick_params(labelbottom = True)
    else:
        ax.tick_params(labelbottom = False)
    
    # Set x, y limits
    ax.set_xlim(1970, 2011)
    ax.set_ylim(0, 100)
    
    # Set y tick values to show only 0 and 100
    ax.set_yticks([0, 100])
    
    # Add a horizontal line at y = 50 position, with a transperancy of 0.3 and a light gray colour in the Colour Blind 10 palette
    ax.axhline(50, c = (171/255, 171/255, 171/255), alpha = 0.3)
    
    # Set title of each plot
    ax.set_title(stem_cats[sp])
    
    # Add text to the plots at given positions
    if sp == 0:
        ax.text(2008, 13, 'Men')
        ax.text(2005, 82, 'Women')
    elif sp == 5:
        ax.text(2008, 85, 'Men')
        ax.text(2005, 10, 'Women')
        
        
# Generate second column of charts. Liberal Arts degrees
for sp in range(0, 5):
    ax = fig.add_subplot(6, 3, 3 * sp + 2)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]], c = cb_dark_blue, linewidth = 3, label = 'Women')
    ax.plot(women_degrees['Year'], 100 - women_degrees[lib_arts_cats[sp]], c = cb_orange, linewidth = 3, label = 'Men')
    
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    
    ax.tick_params(bottom = False, top = False, left = False, right = False)
    
    if sp == 4:
        ax.tick_params(labelbottom = True)
    else:
        ax.tick_params(labelbottom = False)
    
    ax.set_xlim(1970, 2011)
    ax.set_ylim(0, 100)
    ax.set_yticks([0, 100])
    ax.axhline(50, c = (171/255, 171/255, 171/255), alpha = 0.3)
    ax.set_title(lib_arts_cats[sp])

    if sp == 0:
        ax.text(2008, 20, 'Men')
        ax.text(2005, 75, 'Women')
        
# Generate third column of charts. Other degrees
for sp in range(0, 5):
    ax = fig.add_subplot(6, 3, 3 * sp + 3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp]], c = cb_dark_blue, linewidth = 3, label = 'Women')
    ax.plot(women_degrees['Year'], 100 - women_degrees[other_cats[sp]], c = cb_orange, linewidth = 3, label = 'Men')
    
    for key, spine in ax.spines.items():
        spine.set_visible(False)
        
    ax.tick_params(bottom = False, top = False, left = False, right = False)
    
    if sp == 4:
        ax.tick_params(labelbottom = True)
    else:
        ax.tick_params(labelbottom = False)
    
    ax.set_xlim(1970, 2011)
    ax.set_ylim(0, 100)
    ax.set_yticks([0, 100])
    ax.axhline(50, c = (171/255, 171/255, 171/255), alpha = 0.3)
    ax.set_title(other_cats[sp])
    
    if sp == 0:
        ax.text(2008, 5, 'Men')
        ax.text(2005, 88, 'Women')
    elif sp == 5:
        ax.text(2008, 65, 'Men')
        ax.text(2005, 35, 'Women')

# Export the figure containing all the line charts.
plt.savefig('gender_degrees.png')

plt.show()        
```

Conclusions:

STEM:
The Gap between the physical sciences has decreased. Biology was dominated by men, until women closed the gap and eventually surpassed men at what looks like a growing rate. Math and Stats saw a narrow gap in the 90s but that gap has since slightly widened saw a period of nearly no gap until the gap started widening again. Physical sciences saw a 20 year period of narrowing gap before stabilizing.The Gap in the Computer Science field was at its nadir in the mid 80s but has since then only grown wider. The gender gap in Engineering is slowly but surely narrowing.


Liberal Arts:

For nealy all fields in in this category, women constituted the majority. The one except is Social Sciences and History where men are marginally more numerous.  

Others:

Fields such as Health Professions, Public Administration and Education are dominated by women- and continue to be. Women have fully closed the gap in Agriculture and Business. Women have made noticeable inroads in Agriculture, Architecture, and Business. 

With the exception of Computer Science, women are closing the gap in the STEM fields. In other fields, they have either closed the gap or are the dominant group in terms of numbers.
